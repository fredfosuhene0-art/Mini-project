```python
class SmartDevice:

    def __init__(self, name, device_id):
        # Public attribute (no underscore)
        self.name = name
        # Private attributes (encapsulated: because of the double underscore)
        self.__device_id = self._validate_device_id(device_id)
        self.__power_status = False  # power status is false because every device starts from off to on


    # the usb of the @ property
    @staticmethod
    def _validate_device_id(device_id):
        """Ensure that your device ID is never empty."""
        if not device_id or str(device_id).strip() == "":
            raise ValueError("Device ID cannot be empty.")
        return device_id

    # using the Getter / Setter for device_id in order to access and modify the private.
    @property
    def device_id(self):
        """Getter for accessing the private __device_id attribute."""
        return self.__device_id

    @device_id.setter
    def device_id(self, new_id):
        """Setter validates the new ID before allowing a change. used for modifying private __device_id attribute after using the getter"""
        self.__device_id = self._validate_device_id(new_id)

    # using Getter and Setter for power_status
    @property
    def power_status(self):
        """Getter for accessing the private __power_status attribute."""
        return self.__power_status

    @power_status.setter
    def power_status(self, status):
        """Setter only accepts boolean values, protecting internal state. and is used for modification"""
        if not isinstance(status, bool):
            raise ValueError("Power status must be True or False.")
        self.__power_status = status

    # -- Public methods, these methods do not use underscore and can be access directory
    def turn_on(self):
        """When turning the device on."""
        self.power_status = True
        print(f"[{self.name}] is now ON.")

    def turn_off(self):
        """Turning the device off."""
        self.power_status = False
        print(f"[{self.name}] is now OFF.")

    def display_info(self):
        """To display some of the information about the device."""
        status = "ON" if self.power_status else "OFF"
        print("-" * 40)
        print(f"Device Name : {self.name}")
        print(f"Device ID   : {self.device_id}")
        print(f"Power Status: {status}")


# CHILD CLASS for making up SecurityCamera

class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)   # initialize inherited attributes
        self.recording_status = False       # additional attribute only for the security camera

    def start_recording(self):
        """in order to start video recording your device have to be powered on."""
        if not self.power_status:
            print(f"Cannot start recording. Turn on [{self.name}] first.")
            return
        self.recording_status = True
        print(f"[{self.name}] has started recording.")

    def stop_recording(self):
        """To Stop the video recording."""
        self.recording_status = False
        print(f"[{self.name}] has stopped recording.")

    def display_info(self):
        """Extending the parent display_info() with camera-specific details."""
        super().display_info()
        recording = "Recording" if self.recording_status else "Not Recording"
        print(f"Recording Status: {recording}")



# CHILD CLASS 2 for the SmartLight

class SmartLight(SmartDevice):
    """creating a child  class smart light bulb device from the parent class."""

    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__brightness = 0   # for private (self.__brightness) additional attribute for the smart bulb

    @property
    def brightness(self):
        """since self.__brightness is private, getter is used to access it therefore;"""
        return self.__brightness

    @brightness.setter
    def brightness(self, value):
        """using the setter to validates that brightness stays between 0 and 100, this is possible because the getter have already accessed it  """
        if value < 0 or value > 100:
            raise ValueError("Brightness must be between 0 and 100.")
        self.__brightness = value

    def increase_brightness(self, amount=10):
        """ensuring the default brightness is 10."""
        if not self.power_status:
            print(f"Turn on [{self.name}] before adjusting brightness.")
            return
        new_value = min(self.__brightness + amount, 100)
        self.brightness = new_value
        print(f"[{self.name}] brightness increased to {self.brightness}%.")

    def decrease_brightness(self, amount=10):
        """Decreasing brightness by a given amount but the default remains 10 ."""
        if not self.power_status:
            print(f"Turn on [{self.name}] before adjusting brightness.")
            return
        new_value = max(self.__brightness - amount, 0)
        self.brightness = new_value
        print(f"[{self.name}] brightness decreased to {self.brightness}%.")

    def display_info(self):
        """Extending the parent display_info() with the brightness details."""
        super().display_info()
        print(f"Brightness  : {self.brightness}%")



# CHILD CLASS 3 for the TemperatureSensor

class TemperatureSensor(SmartDevice):
    """creating smart temperature sensor device using the details of the mother class."""

    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.__temperature = 0.0   # adding private additional attribute (in Celsius)

    @property
    def temperature(self):
        """using Getter for temperature so that it can be modified by the setter."""
        return self.__temperature

    @temperature.setter
    def temperature(self, value):
        """Modifying temperature using setter."""
        if value < -50 or value > 100:
            raise ValueError("Temperature reading out of realistic range.")
        self.__temperature = value

    def read_temperature(self):

        import random
        if not self.power_status:
            print(f"Turn on [{self.name}] before reading temperature.")
            return
        self.temperature = round(random.uniform(18.0, 32.0), 1)
        print(f"[{self.name}] current temperature: {self.temperature}°C")

    def display_info(self):
        """Extend parent display_info() with temperature details."""
        super().display_info()
        print(f"Temperature : {self.temperature}°C")



# MENU-DRIVEN INTERFACE

def build_devices():
    """Create and return the demo devices required by the project."""
    camera = SecurityCamera("Front Door Camera", "CAM-001")
    light = SmartLight("Living Room Light", "LGT-001")
    sensor = TemperatureSensor("Bedroom Sensor", "TMP-001")
    return {"1": camera, "2": light, "3": sensor}


def print_menu():
    """Display the main menu options."""
    print("\n===== SMART DEVICE MANAGEMENT SYSTEM =====")
    print("1. Display Device Information")
    print("2. Turn Device On")
    print("3. Turn Device Off")
    print("4. Read Temperature")
    print("5. Adjust Brightness")
    print("6. Start Recording")
    print("7. Exit")


def choose_device(devices):
    """Helper function: let the user pick which device to act on."""
    print("\nSelect a device:")
    print("1. Security Camera")
    print("2. Smart Light")
    print("3. Temperature Sensor")
    choice = input("Enter choice (1-3): ").strip()
    return devices.get(choice)


def main():
    """Main program loop implementing the menu-driven interface."""
    devices = build_devices()

    # Demonstrate all methods once automatically before the interactive menu
    print("\n--- AUTOMATIC DEMONSTRATION OF ALL DEVICES/METHODS ---")
    camera, light, sensor = devices["1"], devices["2"], devices["3"]

    for device in (camera, light, sensor):
        device.turn_on()
        device.display_info()

    camera.start_recording()
    camera.stop_recording()

    light.increase_brightness(30)
    light.decrease_brightness(10)

    sensor.read_temperature()

    for device in (camera, light, sensor):
        device.turn_off()

    print("\n--- INTERACTIVE MENU ---")
    while True:
        print_menu()
        option = input("Enter your choice (1-7): ").strip()

        if option == "1":
            device = choose_device(devices)
            if device:
                device.display_info()
            else:
                print("Invalid device selection.")

        elif option == "2":
            device = choose_device(devices)
            if device:
                device.turn_on()
            else:
                print("Invalid device selection.")

        elif option == "3":
            device = choose_device(devices)
            if device:
                device.turn_off()
            else:
                print("Invalid device selection.")

        elif option == "4":
            sensor.read_temperature()

        elif option == "5":
            action = input("Type 'i' to increase or 'd' to decrease brightness: ").strip().lower()
            if action == "i":
                light.increase_brightness()
            elif action == "d":
                light.decrease_brightness()
            else:
                print("Invalid option.")

        elif option == "6":
            action = input("Type 's' to start or 'e' to end recording: ").strip().lower()
            if action == "s":
                camera.start_recording()
            elif action == "e":
                camera.stop_recording()
            else:
                print("Invalid option.")

        elif option == "7":
            print("Exiting Smart Device Management System. Goodbye!")
            break

        else:
            print("Invalid choice. Please select a number from 1 to 7.")


if __name__ == "__main__":
    main()
```
